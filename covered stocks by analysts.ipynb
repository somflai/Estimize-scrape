## Json code

```python
for i in range(len(ana)):
    lnk ='https://www.estimize.com/users/'+str(analyst[i])
    browser.get(lnk)
    
    datainfor = {}
    try:
        browser.execute_script("window.scrollTo(0, 900)")
        WebDriverWait(browser,20).until(EC.presence_of_element_located((By.XPATH, '//*[@id="profile-covered-stocks"]/div[1]')))
        browser.find_element_by_xpath('//*[@id="profile-covered-stocks"]/div[2]/a').click()
        data1[i] = datainfor
        
    except:
        
        continue
    for k in range(1,100):
                                                                   
        try:
            datas = {}
            covered ='//*[@id="profile-covered-stocks"]/div[1]/div[2]/div['+str(k)+']/div[1]/a'
            quarters ='//*[@id="profile-covered-stocks"]/div[1]/div[2]/div['+str(k)+']/div[3]'
            points ='//*[@id="profile-covered-stocks"]/div[1]/div[2]/div['+str(k)+']/div[4]'
            pest ='//*[@id="profile-covered-stocks"]/div[1]/div[2]/div['+str(k)+']/div[5]'
            error ='//*[@id="profile-covered-stocks"]/div[1]/div[2]/div['+str(k)+']/div[6]'
            accuracy ='//*[@id="profile-covered-stocks"]/div[1]/div[2]/div['+str(k)+']/div[7]'

            datas['analyst'] = analyst[i]
            datas['stocks_covered'] = browser.find_element_by_xpath(covered).text
            datas['stocks_quarters'] = browser.find_element_by_xpath(quarters).text
            datas['stocks_points'] = browser.find_element_by_xpath(points).text
            datas['stocks_pte_per_est'] = browser.find_element_by_xpath(pest).text
            datas['stocks_errorrate'] = browser.find_element_by_xpath(error).text
            datas['stocks_accuracy'] = browser.find_element_by_xpath(accuracy).text
            print(datas)
            datainfor[k] = datas
            
        except:
            
            continue
            
    
            
    
```
